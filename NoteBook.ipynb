Analysis

```
# split rwandan data
r_data = df_2.Country.str.contains('Rwanda')
rwandan = df_2[r_data]
rwandan
```

```
# total tax
t_tax = rwandan['Total Taxes'].astype('float')
print(t_tax.max())
```

```
# which year had the most tax
maxValue = rwandan[['Year', 'Total Taxes']].astype('float')
print("Maximum value in column 'y' & 'z': ")
print(maxValue.max())
```

```
# checking for inflation and taxes collected
t_tax = rwandan['Inflation\n(Av CPI)'].astype('float')
print(t_tax.max())
```

```
maxtax = rwandan[['Direct Taxes', 'Corporate Tax','Individual Tax','Indirect Taxes','Trade Taxes','Resource Taxes','Non-Resource Taxes']].max()
print("Maximum taxes")
print(maxtax)
```

```
maxtax.plot.pie(figsize=(7, 7))
```

```
# country that collected the maximum tax and the year
maxcont= df_na[['Country', 'Total Taxes']].max()
print("Maximum value in column 'Country' & 'Total Taxes': ")
print(maxcont)
```

```
# split to South Africa
s_data = df_2.Country.str.contains('South Africa')
south_a = df_2[s_data]
south_a
```

```
# Comparisson to South Africa
maxtax_sa = south_a[['Direct Taxes', 'Corporate Tax','Individual Tax','Indirect Taxes','Trade Taxes','Resource Taxes','Non-Resource Taxes']].max()
print("Maximum taxes")
print(maxtax_sa)
```

```
# visualization of high taxation
maxtax_sa.plot.bar()
```

```
#  south african inflation
# checking for inflation and taxes collected
t_tax = south_a['Inflation\n(Av CPI)'].astype('float')
print(t_tax.max())
```
